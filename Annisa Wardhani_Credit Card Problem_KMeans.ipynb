# 1. Business Problem Understanding

### Context
Berdasarkan Undang-Undang Nomor 10 Tahun 1998 tentang Perbankan, **bank** adalah badan usaha yang menghimpun dana dari masyarakat dalam bentuk simpanan dan menyalurkannya kepada masyarakat dalam bentuk kredit dan atau bentuk-bentuk lainnya dalam rangka meningkatkan taraf hidup masysarakat. Bank berperan dalam menghimpun dana, menyalurkan dana, dan memberikan jasa-jasa bank lain. Bank menyediakan berbagai produk untuk nasabahnya seperti rekening giro, tabungan, deposito, kredit, kartu kredit, asuransi, layanan perbankan elektronik, investasi dan manajemen kekayaan. Seperti bank pada umumnya, bank di Amerika menyediakan deposito, kredit, layanan pembayaran (kartu debit dan kartu kredit), perencanaan keuangan, investasi, dan manajemen kekayaan.

Salah satu produk di dunia perbankan adalah kartu kredit.  Kartu kredit adalah produk pembayaran untuk nasabah dan transaksi bisnis. Berdasarkan **[Trading Economics](https://id.tradingeconomics.com/united-states/credit-card-accounts)** jumlah akun kartu kredit di Amerika Serikat mengalami kenaikan pada tahun 2023. Hampir setiap penduduk di Amerika dari berbagai golongan mendapatkan tawaran pembuatan kartu kredit melalui telepon atau *email* setiap minggunya. Nasabah kartu kredit dan bank akan memdapatkan keuntungan apabila sang nasabah sering melakukan transaksi menggunakan kartu kredit dan tetap menyimpan kartu kreditnya dalam jangka waktu yang lama. Selain itu, pihak bank akan mendapatkan keuntungan besar.

Bank Bangtut adalah suatu bank di Amerika. Bank Bangtut hanya memiliki satu program produk kartu kredit saja. Bank bangtut ingin menambah keuntungan bank lebih besar melalui produk kartu kredit dengan cara melakukan segmentasi atau menganalisa nasabah kartu kredit. Segmentasi nasabah kartu kredit dapat dilakukan dengan cara mengelompokkan para nasabah berdasarkan profil dan kebiasaan dari masing-masing nasabah agar Bank Bangtut menyusun strategi bisnis yang lebih efektif dan optimal, dan juga dapat memberikan *treatment* yang tepat untuk masing-masing nasabah kartu kredit.


### Problem Statement
Setiap bank memiliki produk kartu kredit sehingga satu tantangan dari Bank Bangtut adalah mempertahankan nasabah kartu kredit agar tetap menggunakan kartu kredit dari Bank Bangtut. Berdasarkan dari **[FPSC](https://www.fpsc.com/the_cost_of_customer_churn.pdf)** *cost* mendapatkan nasabah baru **lima kali** lebih besar dibandingkan mempertahankan yang sudah ada. Dengan mengetahui hal tersebut, itu menjadi tantangan tersendiri untuk bank Bangtut. Selama ini belum ada *treatment* yang diberikan kepada nasabah secara spesifik melihat profil-profil nasabah. Oleh sebab itu, masih ada permasalah untuk memberikan *treatment* ke masing-masing nasabah agar mereka tetap berlangganan kartu kredit di bank Bangtut dan sering bertransaksi menggunakan kartu kreditnya.

Kami sebagai tim data bekerja sama dengan tim *marketing* hendak melakukan sebuah strategi pemasaran dan *campaign* dalam bisnis dengan tujuan untuk mengelompokkan semua nasabah kartu kredit menjadi beberapa kelompok berdasarkan profil dan kebiasaan yang sama. Tujuannya adalah *treatment* atau strategi pemasaran yang diberikan kepada para nasabah tepat sesuai dengan target dan meminimalkan *cost* (*cost effectivity*) sehingga para nasabah akan sering menggunakan kartu kreditnya dan tetap berlangganan kartu kredit di Bank Bangtut.


### Goals
Berdasarkan dari permasalahan di atas, tujuan kami yaitu dapat menjabarkan karakteristik nasabah kartu kredit saat ini dan selanjutnya akan kami kelompokkan berdasarkan riwayat transaksi kartu kreditnya. Kami juga akan memberikan saran untuk tim *marketing* mengenai strategi pemasaran atau *tretment* yang tepat untuk diberikan ke masing-masing nasabah sehingga mereka tetap menggunakan kartu kredit bank Bangtut dan meningkatkan transaksi. Akhirnya, bank akan mendapatkan profit.

Sebelum memulai EDA, kita lihat kembali bentuk dan info dataset baru

Dari informasi di atas, kita bisa mendapatkan dua informasi penting yaitu:
* `LIMIT_USE`: rasio penggunaan limit kartu kredit berdasarkan saldo yang telah digunakan dibandingkan dengan limit kredit yang tersedia. Tujuannya untuk mengukur sejauh mana nasabah telah menggunakan batas kredit yang disetujui oleh bank. Dengan kata lain berhubungan dengan *credit score* 
`LIMIT_USE` = `BALANCE` / `CREDIT_LIMIT`    
Semakin tinggi limit use, semakin dekat nasabah dengan batas kredit mereka, yang dapat mempengaruhi profil risiko kredit mereka.

* `PAYMENT_MINPAYMENT`: perbandingan antara jumlah pembayaran yang dilakukan oleh nasabah terhadap jumlah pembayaran minimum yang harus dibayarkan setiap bulan. Rasio ini penting karena memberikan gambaran tentang sejauh mana nasabah melunasi kewajiban pembayaran minimum mereka.  
`PAYMENT_MINPAYMENT` = `PAYMENT` / `MINIMUM_PAYMENTS`

```python
# menambahkan kolom LIMIT_USE dan PAYMENT_MINPAYMENT
df['LIMIT_USE']= df['BALANCE']/df['CREDIT_LIMIT']
df['PAYMENT_MINPAYMENT']=df['PAYMENTS']/df['MINIMUM_PAYMENTS']
```

Berdasarkan fitur dataset yang terdiri dari `PURCHASES`, `ONEOFF_PURCHASES`, dan `INSTALLMENTS_PURCHASES`, nasabah kartu kredit Bank Bangtut dapat dikelompokkan ke dalam empat profil pembelian yang berbeda:

1. **Oneoff Purchases Only**: Kelompok ini terdiri dari nasabah yang hanya melakukan pembelian sekali atau pembelian tanpa cicilan saja.

2. **Installment Purchases Only**: Kelompok ini terdiri dari nasabah yang hanya melakukan pembelian dengan sistem cicilan saja.

3. **Both Purchases**: Kelompok ini mencakup nasabah yang melakukan pembelian tanpa cicilan dan juga pembelian dengan cicilan.

4. **None**: Kelompok ini mencakup nasabah yang tidak melakukan pembelian.

Dengan mengkategorikan nasabah ke dalam empat profil pembelian ini berdasarkan perilaku pembelian mereka, Bank Bangtut dapat memperoleh wawasan berharga tentang preferensi nasabah, kebiasaan finansial, serta area potensial untuk strategi pemasaran yang lebih terarah atau penanganan yang disesuaikan.

```python
# menambahkan kolom PURCHASES TYPE
df.loc[(df.ONEOFF_PURCHASES==0) & (df.INSTALLMENTS_PURCHASES==0),'PURCHASES_TYPE']='None'
df.loc[(df.ONEOFF_PURCHASES>0) & (df.INSTALLMENTS_PURCHASES==0) , 'PURCHASES_TYPE']= 'Oneoff Only'
df.loc[(df.ONEOFF_PURCHASES==0) & (df.INSTALLMENTS_PURCHASES>0) , 'PURCHASES_TYPE']= 'Installment Only'
df.loc[(df.ONEOFF_PURCHASES>0) & (df.INSTALLMENTS_PURCHASES>0), 'PURCHASES_TYPE']= 'Both'

df.head()
```

# 6. Modeling

## 6.1. K-Means Clustering

### 6.1.1. Metric Evaluation

***Elbow Method***

*Elbow Method* merupakan salah satu metode yang digunakan untuk menentukan banyaknya cluster yang optimal dalam algoritma K-Means.

```python
# Optimum elbow score
wcss = []
for i in range(3, 16):
    kmeans_cancer=KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_cancer.fit(dff_log_scaled)
    wcss.append(kmeans_cancer.inertia_)
plt.plot(range(3, 16), wcss, marker = 'o')
plt.title('Elbow Method')
plt.xlabel('Jumlah Cluster')
plt.ylabel('Average Within Sum of Square')
plt.show()
```

Jadi kalau dilihat dari grafik yang sudah kita peroleh, terlihat bahwa terdapat beberapa kemungkinan yang menjadi *cluster* terbaik, diantara *cluster* dengan jumlah diantara 4-6 *cluster*. Setelah *cluster* 7 penurunan terjadi lebih konstan yaitu dibawah 1k WCSS dan terlihat lebih landai.

Namun, perlu dicatat bahwa penggunaan metode ***Elbow Method*** terkadang dapat bersifat subjektif, sehingga diperlukan validasi menggunakan metode lain, seperti ***silhouette scores***.

***Silhoutte Score***

*Silhouette score* adalah metode evaluasi *clustering* yang digunakan untuk menentukan seberapa baik suatu data dikelompokkan dalam suatu *cluster*. *Silhouette score* menghitung seberapa dekat setiap titik data dalam suatu *cluster* dengan titik data di cluster lain dibandingkan dengan titik data di cluster yang sama. *Silhouette Score* berkisar antara -1 hingga 1, dengan nilai yang lebih tinggi menunjukkan cluster yang lebih baik. Semakin dekat nilai *silhouette score* dengan 1, semakin baik data dikelompokkan.

```python
# Optimum silhouette score
SC = []
cluster_range = range(4, 11)
for i in cluster_range:
    kmeans_boston = KMeans(n_clusters=i)
    kmeans_boston.fit(dff_log_scaled)
    labels = kmeans_boston.labels_
    silhouette_avg = silhouette_score(dff_log_scaled, labels, metric='euclidean')
    SC.append(silhouette_avg)

plt.figure(figsize=(8, 4))
sns.lineplot(x=cluster_range, y=SC)
sns.scatterplot(x=cluster_range, y=SC)

plt.xticks(cluster_range)
plt.xlabel('Jumlah Cluster')
plt.ylabel('Silhouette Score')

plt.show()
```

Jika diperhatikan pada gambar diatas, terlihat bahwa urutan yang memiliki silhoute score terbesar adalah sebagai berikut:
- 4 Cluster dengan *silhoutte score* sekitar 3.4
- 5 Cluster dengan *silhoutte score* sekitar 3.6

### 6.1.2. Model Clustering

Dalam tahap ini kita akan melakukan *clustering* dengan metode **K-Means** dan hasil pengelompokkan akan divisualisasikan dengan metode **PCA** pada masing-masiing pembagian.

Tujuan penggunaan metode **PCA** adalah untuk memudahkan kita dalam mevisualisasikan hasil *clustering*.

Metode **PCA (*Principal Component Analysis*)** adalah teknik statistik yang digunakan untuk mengurangi dimensi data dengan tujuan mempertahankan sebanyak mungkin informasi varian dalam data yang ada. **PCA** mengubah data yang mungkin berkorelasi menjadi serangkaian variabel baru yang disebut ruang komponen utama yang lebih rendah. Komponen utama ini adalah kombinasi linear dari variabel asli yang saling tidak berkorelasi.  

**PCA** dapat membantu dengan mengurangi dimensi data agar lebih mudah dipahami dan diproses oleh algoritma *clustering*. Dengan mengurangi dimensi data, **PCA** dapat membantu mengatasi masalah kutukan dimensi *(curse of dimensionality)* yang dapat mempengaruhi kinerja algoritma *clustering*.

Untuk kasus kali ini, kita hanya menggunaka **2** komponen utama saja (**PC1** dan **PC2**). Hanya menggunakan **2** komponen utama sudah cukup untuk memvisualisasikan data secara efektif.  Komponen utama pertama (**PC1**) menjelaskan variasi terbesar dalam data, sedangkan komponen utama kedua (**PC2**) menyediakan variasi tambahan yang saling ortogonal terhadap **PC1**.  

Dari **2** komponen utama tersebut, kita dapat memvisualisasikan data dalam bentuk scatter plot dua dimensi (sumbu X dan Y). Sumbu X dan Y dalam konteks **PCA** merepresentasikan sumbu-sumbu baru dalam ruang komponen utama setelah proyeksi data. Sumbu X (sumbu horizontal) dan Sumbu Y (sumbu vertikal) adalah dua sumbu utama yang terbentuk setelah reduksi dimensi menggunakan PCA. Masing-masing sumbu ini merupakan kombinasi linier dari variabel asli dalam data, di mana variabel-variabel tersebut memberikan kontribusi terhadap arah sumbu tersebut sesuai dengan bobot (beban) mereka pada komponen utama. Sumbu X dan Y memungkinkan kita untuk memvisualisasikan data dalam ruang dua dimensi, dengan setiap titik data direpresentasikan oleh koordinat pada sumbu X dan Y. Visualisasi ini dapat membantu kita memahami pola, kelompok, atau hubungan antara data dengan cara yang lebih intuitif.
